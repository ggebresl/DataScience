## Segment 2 - Defining elements of a plot - Defining axes, ticks, and grids 

```python
x = range(1,10) 

y = [1,2,3,4,0,4,3,2,1] 
fig = plt.figure() 
ax = fig.add_axes([.1, .1, 1, 1]) 
ax.plot(x,y)

#ax2 = fig.add_axes([.2, .2, 2, 2])
#ax2.plot(x,y)


```

```python
fig = plt.figure() 
ax = fig.add_axes([.1, .1, 1, 1])  

ax.set_xlim([1,9]) 
ax.set_ylim([0,5]) 
ax.set_xticks([0,1,2,4,5,6,8,9,10]) 
ax.set_yticks([0,1,2,3,4,5]) 
ax.plot(x,y) 
```

```python
fig = plt.figure() 
ax = fig.add_axes([.1, .1, 1, 1])  
ax.set_xlim([1,9]) 
ax.set_ylim([0,5]) 
ax.grid() 
ax.plot(x, y) 

```

## Generating multiple plots in one figure with subplots 


```python
fig = plt.figure() 
fig, (ax1, ax2) = plt.subplots(1,2) 
ax1.plot(x) 
ax2.plot(x,y) 

```

## Segment 4 - Creating labels and annotations 

```python
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame 
import matplotlib.pyplot as plt 
from pylab import rcParams 
import seaborn as sb 
%matplotlib inline 
rcParams['figure.figsize'] = 8,4 
sb.set_style('whitegrid') 

```

## Labeling plot features The functional method 

```python
x =  range(1,10) 
y = [1,2,3,4,0.5,4,3,2,1] 
plt.bar(x,y) 
plt.xlabel('your x-axis label') 
plt.ylabel('your y-axis label') 
```

```python
z = [1 , 2, 3, 4, 0.5] 
veh_type = ['bicycle', 'motorbike','car', 'van', 'stroller'] 
plt.pie(z, labels= veh_type) 
plt.show() 
```

## The object-oriented method 

```python
address = 'dataset/mtcars.csv' 
cars = pd.read_csv(address) 
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb'] 
```

```python
mpg = cars.mpg 
fig = plt.figure() 

ax = fig.add_axes([.1, .1, 1, 1])
mpg.plot() 
ax.set_xticks(range(32)) 
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium') 
ax.set_title('Miles per Gallon of Cars in mtcars') 
ax.set_xlabel('car names') 
ax.set_ylabel('miles/gal') 


```

## Adding a legend to your plot The functional method 

```python
plt.pie(z) 
plt.legend(veh_type, loc='best') 
plt.show() 
```

## The object-oriented method 

```python
fig = plt.figure() 
ax = fig.add_axes([.1,.1,1,1]) 

mpg.plot() 
ax.set_xticks(range(32)) 
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium') 
ax.set_title('Miles per Gallon of Cars in mtcars') 
ax.set_xlabel('car names') 
ax.set_ylabel('miles/gal') 
ax.legend(loc='best') 
```

## Annotating your plot 

```python
print(mpg.max()) 
```

```python
fig = plt.figure() 
ax = fig.add_axes([.1,.1,1,1]) 

mpg.plot() 
ax.set_title('Miles per Gallon of Cars in mtcars') 
ax.set_ylabel('miles/gal') 
ax.set_ylim([0,45]) 
ax.annotate('Toyota Corolla', xy=(19,33.9), 
xytext = (21,35), arrowprops=dict(facecolor='black', shrink=0.05)) 
```

## Segment 6 - Constructing histograms, box plots, and scatter plots 

```python
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame 
from pandas.tools.plotting import scatter_matrix 
import matplotlib.pyplot as plt 
from pylab import rcParams 
import seaborn as sb 




```

```python
%matplotlib inline 
rcParams['figure.figsize'] = 5, 4 
sb.set_style('whitegrid') 
```

## Eyeballing dataset distributions with histograms 

```python
address = 'dataset/mtcars.csv' 
cars = pd.read_csv(address) 
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb'] 
cars.index = cars.car_names 
mpg = cars['mpg'] 
mpg.plot(kind='hist') 
```

```python
plt.hist(mpg) 
plt.plot() 
```

## Seeing scatterplots in action 

cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150) 

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True) 
```

## Generating a scatter plot matrix 

```python
sb.pairplot(cars) 
```

```python


cars_df = pd.DataFrame((cars.ix[:,(1,3,4,6)].values), columns = ['mpg', 'disp', 'hp', 'wt']) 
cars_target = cars.ix[:,9].values 
target_names = [0, 1] 
cars_df['group'] = pd.Series(cars_target, dtype="category") 
sb.pairplot(cars_df, hue='group', palette='hls') 
```

## Building boxplots 

```python
cars.boxplot(column='mpg', by='am') 
cars.boxplot(column='wt', by='am') 
```
