```python
import pandas as pd
import numpy as np
import seaborn as sb
filename = '../7_Matplotlib/datasets/mtcars.csv'
cars = pd.read_csv(filename)
cars.head()
```

```python
cars_cat = cars[['cyl','vs','am','gear','carb']]
print(cars_cat.head())
gear_group = cars_cat.groupby('gear')
gear_group.describe()
```

```python
cars.quantile([0.25,0.50,0.75])
```

```python
pd.crosstab(cars['am'],cars['gear'])
```

```python
sb.pairplot(cars)
```
